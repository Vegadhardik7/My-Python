```python
from sklearn.preprocessing import OrdinalEncoder  
oe = OrdinalEncoder(categories=[['low','med','high','vhigh'], 
                                ['low','med','high','vhigh'], 
                                ['2','3','4','5more'], 
                                ['2','4','more'], 
                                ['small','med','big'], 
                                ['low','med','high'], ['unacc','acc','good','vgood']])

oe.fit_transform(data[['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'Class_Distribution']]) 
```

```python
data[['buying', 
      'maint', 
      'doors', 
      'persons', 
      'lug_boot', 
      'safety', 
      'Class_Distribution']] = oe.fit_transform(data[['buying', 'maint', 'doors', 
                                                      'persons', 'lug_boot', 'safety', 'Class_Distribution']])
```

```python
data.head()
```

```python
# This will give the correlation between that features

# Example : age in Y-axis and age in X-axis will have the maximum correlation that's why it's dark in color

plt.figure(figsize=(24,15))                                       # Help to set figure size                                     

sns.heatmap(data.corr(),annot=True,cmap='hsv',fmt='.5f',linewidths=20)    
         
# annot : Help to see the value in box
# cmap : The mapping from data values to color space.
# fmt : String formatting code to use when adding annotations.

plt.ylim(15,0)                                                     # show us exact number of values we want
plt.show()
```

```python
X = data.drop('Class_Distribution', axis=1)
y = data['Class_Distribution']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=8)

model.fit(X_train, y_train)

acc = model.score(X_test, y_test)
print(acc)
```

```python
plt.plot(X_train, y_train)
```

```python
predicted = model.predict(X_test)

val = pd.DataFrame(np.c_[y_test, predicted], columns=['Actual','Predicted'])
```
